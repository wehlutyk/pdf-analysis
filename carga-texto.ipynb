```python
import os
import os.path
```

```python
textdir = '/home/sl/Code/Research/pdf-analysis-unc/text/'
```

```python
n_articles = 0
journals = []
titles = []
contents = []

for journal in os.listdir(textdir):
    print "Processing articles from '{}'".format(journal.title())
    journal_path = os.path.join(textdir, journal)
    for article in os.listdir(journal_path):
        n_articles += 1
        journals.append(journal)
        titles.append(article)

        article_path = os.path.join(journal_path, article)
        with open(article_path, 'r') as a:
            contents.append("".join(a.readlines()))

print
print "{} articles processed.".format(n_articles)
```
